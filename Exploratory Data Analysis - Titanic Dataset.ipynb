```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv(r"C:\Users\Loges\Downloads\Datasets\titanic\train.csv")
data
```

```python
data['Sex']=data['Sex'].replace({'male':0,'female':1})
data['Embarked']=data['Embarked'].replace({np.nan:0,'S':1,'C':2,'Q':3}).astype('int64')
```

```python
data=data.drop(["Cabin","Name","Ticket"],axis=1)
data
```

```python
data.corr()
```

```python
sns.heatmap(data.corr(), annot=True)
```

```python
def compare(col1,col2,data):
    count=[[0 for i in data[col1].unique()] for j in data[col2].unique()]
    for i in range(len(data)):
        count[data[col2][i]][data[col1][i]]+=1
    return count
```

```python
plt.figure(figsize=(20,5))

count, col1, col2=compare("Survived","Sex",data),["No","Yes"],["Male","Female"]
plt.subplot(1,2,1)
sns.heatmap(count,annot=count,xticklabels=col1,yticklabels=col2,fmt="5d")
plt.ylabel("Sex")
plt.xlabel("Survived")

plt.subplot(1,2,2)
count=compare("Sex","Survived",data)
n=np.array([0,1])
plt.bar(n-0.25, count[0], width=0.25)
plt.bar(n, count[1], width=0.25)
plt.xticks(n-0.125,["Male","Female"])
plt.xlabel("Sex")
plt.ylabel("Count")
plt.legend(["Not Survived","Survived"])
plt.show()
```

```python
plt.figure(figsize=(20,5))
count, col2, col1=compare("Parch","Survived",data),["No","Yes"],[0,1,2,3,4,5,6]
plt.subplot(1,2,1)
sns.heatmap(count,annot=count,xticklabels=col1,yticklabels=col2,fmt="5d")
plt.xlabel("Parch")
plt.ylabel("Survived")

plt.subplot(1,2,2)
n=np.array(col1)
plt.bar(n-0.25, count[0], width=0.25)
plt.bar(n, count[1], width=0.25)
plt.xticks(n-0.125,col1)
plt.xlabel("Parch")
plt.ylabel("Count")
plt.legend(["Not Survived","Survived"])
```

```python
plt.figure(figsize=(20,5))
count,col2,col1=compare("Embarked","Survived",data),["No","Yes"],["Null",'S','C','Q']
plt.subplot(1,2,1)
sns.heatmap(count,annot=count,xticklabels=col1,yticklabels=col2,fmt="5d")
plt.xlabel("Embarked Class")
plt.ylabel("Survived")

plt.subplot(1,2,2)
n=np.array([0,1,2,3])
plt.bar(n-0.25, count[0], width=0.25)
plt.bar(n, count[1], width=0.25)
plt.xticks(n-0.125,col1)
plt.xlabel("Embarked")
plt.ylabel("Count")
plt.legend(["Not Survived","Survived"])
plt.show()
```
